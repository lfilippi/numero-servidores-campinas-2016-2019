```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline

pd.set_option('precision', 1)
```

```python
codigos_cargos = pd.read_csv('codigos_cargos.csv', encoding="utf-8")
funcionalismo_janeiro_2016 =  pd.read_csv('01_janeiro_2016.csv', encoding="utf-8")
funcionalismo_janeiro_2016['Código Cargo'] = funcionalismo_janeiro_2016['Código Cargo'].astype('int64') 

funcionalismo_julho_2019 =  pd.read_csv('07_julho_2019.csv', encoding="utf-8")
funcionalismo_julho_2019['Código Cargo'] = funcionalismo_julho_2019['Código Cargo'].astype('int64') 
```

```python
funcionalismo_janeiro_2016 = funcionalismo_janeiro_2016.groupby(['Código Cargo', 'Nome Cargo']) 
funcionalismo_janeiro_2016 = funcionalismo_janeiro_2016.size()
funcionalismo_janeiro_2016 = funcionalismo_janeiro_2016.reset_index()
funcionalismo_janeiro_2016.columns = ['Código', 'Nome', 'Janeiro 2016'] # change column names

funcionalismo_julho_2019 = funcionalismo_julho_2019.groupby(['Código Cargo', 'Nome Cargo']) 
funcionalismo_julho_2019 = funcionalismo_julho_2019.size()
funcionalismo_julho_2019 = funcionalismo_julho_2019.reset_index()
funcionalismo_julho_2019.columns = ['Código', 'Nome', 'Julho 2019'] # change column names

funcionalismo_janeiro_2016.head()
```

```python
comparativo_2016_2019 = funcionalismo_janeiro_2016.merge(funcionalismo_julho_2019, how='outer', on='Código')
comparativo_2016_2019['Código'] = comparativo_2016_2019['Código'].astype('int64') 
comparativo_2016_2019 = comparativo_2016_2019.merge(codigos_cargos, how='outer', on=['Código'])
comparativo_2016_2019 = comparativo_2016_2019.set_index('Denominação dos cargos')
comparativo_2016_2019 = comparativo_2016_2019.drop(['Grupo', 'Nome_x', 'Nome_y', 'Comissionado', 'Subgrupo (Educação)', 'Posição (Sinfônica)', 'Especialidade (Sinfônica)'], axis=1)
```
